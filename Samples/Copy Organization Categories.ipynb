### Establish your GIS Connections

```python
import arcgis
src_gis = arcgis.gis.GIS("https://org1.maps.arcgis.com","user1")
print(src_gis._username, "@", src_gis)
target_gis = arcgis.gis.GIS("https://org2.maps.arcgis.com", "user2")
print(target_gis._username, "@", target_gis)
```

#### Display the current categories for your organizations

```python
src_categories = src_gis.content.categories.schema
print ("Source Categories:")
import pprint
pprint.pprint(src_categories)
```

#### Now, apply your Source Organization's Categories to your Target Organization. We will preserve the Original Target Categories if needed in this notebook.

```python
target_categories_bak = target_gis.content.categories.schema
target_gis.content.categories.schema = src_categories
print("New Categories:\n~~~~~~~~~~~~~")
pprint.pprint(target_gis.content.categories.schema)
print("\nOriginal Categories\n~~~~~~~~~~~~")
pprint.pprint(target_categories_bak)
```
